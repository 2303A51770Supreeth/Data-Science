```python
import numpy as np
import matplotlib.pyplot as plt

temps = np.array([15.2, 17.8, 22.4, 28.9, 32.0, 34.5, 33.1, 32.8, 30.0, 26.1, 21.0, 17.0])

avg_temp = np.mean(temps)

hottest_idx = np.argmax(temps)
coldest_idx = np.argmin(temps)

normalized = (temps - np.min(temps)) / (np.max(temps) - np.min(temps))

months = np.arange(1, 13)
above_avg_months = months[temps > avg_temp]

fahrenheit = temps * 9/5 + 32

print("Monthly Temperatures (°C):", temps)
print("Average Temperature:", round(avg_temp, 2))
print("Hottest Month:", hottest_idx+1, "Temp:", temps[hottest_idx])
print("Coldest Month:", coldest_idx+1, "Temp:", temps[coldest_idx])
print("Normalized Values:", np.round(normalized, 2))
print("Months Above Average:", above_avg_months)
print("Temperatures in °F:", np.round(fahrenheit, 2))

plt.figure(figsize=(8,4))
plt.plot(months, temps, marker='o', label='°C')
plt.plot(months, fahrenheit, marker='s', label='°F', linestyle='--')
plt.axhline(avg_temp, color='red', linestyle=':', label='Avg Temp (°C)')
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Headphones"],
    "Quantity": [2, 5, 1, 3, 10, 4],
    "Price": [50000, 15000, 50000, 20000, 15000, 3000]
}

df = pd.DataFrame(data)

df["TotalSale"] = df["Quantity"] * df["Price"]

high_value_orders = df[df["TotalSale"] > 1000]

product_sales = df.groupby("Product")["Quantity"].sum()
most_sold_product = product_sales.idxmax()

df["Discount"] = np.where(df["TotalSale"] > 50000, df["TotalSale"] * 0.10, df["TotalSale"] * 0.05)

df.to_excel("sales_summary.xlsx", index=False)

print("Original DataFrame:\n", df)
print("\nHigh Value Orders:\n", high_value_orders)
print("\nProduct Sales Summary:\n", product_sales)
print("\nMost Sold Product:", most_sold_product)
print("\nExcel file 'sales_summary.xlsx' saved. Use left panel in Colab to download.")
```
